```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop([ 'class', 'who', 'adult_male', 'deck', 'embark_town',
       'alive'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.dropna(subset=['embarked'],inplace=True)
```

```python
#for label encodeing
from sklearn.preprocessing import LabelEncoder 
```

```python
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df.head()
```

```python
df=df.astype(int)
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.33, random_state=42)
```

LOGISITIC REGRESSION

```python
#for logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
#0.8163265306122449
```

```python
confusion_matrix(y_test,y_pred)
"""array([[157,  27],
       [ 27,  83]])"""
```

```python
print(classification_report(y_test,y_pred))
```

KNN

```python
# k nearest negibhour
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train_scaled=ss.fit_transform(x_train)
x_test_scaled=ss.fit_transform(x_test)
x_test_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier
kmodel=KNeighborsClassifier(n_neighbors=5)
kmodel.fit(x_train_scaled,y_train)
```

```python
y_pred_knn=kmodel.predict(x_test_scaled)
accuracy_score(y_test,y_pred_knn)
#0.8061224489795918
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print(classification_report(y_test,y_pred_knn))
```

NAVIE BAYES

```python
#by navie bayes
from sklearn.naive_bayes import GaussianNB
NB=GaussianNB()

```

```python
NB.fit(x_train,y_train)
```

```python
y_predict_nb=NB.predict(x_test)
```

```python
y_predict_nb
```

```python
y_test
```

```python
accuracy_score(y_test,y_predict_nb)
#0.778911564625850
```

```python
print(classification_report(y_test,y_predict_nb))
```

DECISION TREE

```python
#decision tree 
#it needed scaling
from sklearn.tree import DecisionTreeClassifier
model_DE=DecisionTreeClassifier(random_state=42)

```

```python
model_DE.fit(x_train_scaled,y_train)
```

```python
y_pred_DE=model_DE.predict(x_test_scaled)
y_pred_DE
```

```python
y_test
```

```python
accuracy_score(y_pred_DE,y_test)
#0.789115646258503
```

```python
print(classification_report(y_test,y_pred_DE))
```

```python
confusion_matrix(y_pred_DE,y_test)
```

SVM

```python
#by SVM
#needed feture scaling

from sklearn.svm import SVC
model_SVM=SVC(kernel='rbf')
```

```python
model_SVM.fit(x_train_scaled,y_train)
```

```python
y_pred_svm=model_SVM.predict(x_test_scaled)
```

```python
accuracy_score(y_test,y_pred_svm)
#0.8197278911564626  NOW THIS IS THE BEST MODEL
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
print(classification_report(y_test,y_pred_svm))
```
