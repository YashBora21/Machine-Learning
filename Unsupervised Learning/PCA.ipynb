```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
x,y=make_blobs(n_samples=500, centers=3, n_features=5,cluster_std=1.5,random_state=42)
```

```python
x
```

```python
df=pd.DataFrame(x,columns=['f1','f2','f3','f4','f5'])
```

```python
df
```

```python
scaler=StandardScaler()
x_sclaed=scaler.fit_transform(x)
```

```python
x_sclaed
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_sclaed
                        )
```

```python
x_pca
```

```python
df_pca=pd.DataFrame(x_pca,columns=['pca1','pca2'])
df_pca['label']=y
```

```python
sns.scatterplot(x=df_pca['pca1'],y=df_pca['pca2'],hue=df_pca['label'],palette='rainbow')
```
