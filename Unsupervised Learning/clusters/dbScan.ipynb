```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN,KMeans
from sklearn.preprocessing import StandardScaler
```

```python
x,y=make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
x
```

```python
df=pd.DataFrame(x,columns=['features1','features2'])
```

```python
Scaler=StandardScaler()
x_scaled=Scaler.fit_transform(df)
```

FIRST WE WILL SEE IS KMEANS CLUSTRING WORKS?

```python
kmeans_final=KMeans(n_clusters=3,random_state=42)
cluster_labels=kmeans_final.fit_predict(x_scaled)
```

```python
df['label']=cluster_labels
```

```python
sns.scatterplot(x=df['features1'],y=df['features2'],hue=df['label'],palette='rainbow')
```

```python
dbscan=DBSCAN(eps=0.3, min_samples=5)
db_label=dbscan.fit_predict(x_scaled)
```

```python
df['lable1']=db_label
df.head()
```

```python
sns.scatterplot(x=df['features1'],y=df['features2'],hue=df['lable1'],palette='rainbow')
```
