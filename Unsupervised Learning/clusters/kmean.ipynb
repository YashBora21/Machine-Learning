```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs #to make the data sets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
x,ytrue=make_blobs(n_samples=500, centers=3,cluster_std=0.60,random_state=42)
```

```python
x
```

```python
df=pd.DataFrame(x,columns=['features1','features2'])
```

```python
df
```

```python
scaler= StandardScaler()
x_scaled=scaler.fit_transform(df)
```

```python
#elbpw method
wcss=[]
k_range=range(1,11)
```

```python
for k in k_range:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
"""[1000.0,
 297.8954141051723,
 11.57548472310498,
 9.752067977356841,
 8.257175272446284,
 6.9175773204168,
 6.334755391595291,
 5.704177177901429,
 5.060234133532076,
 4.762361898130396]"""
```

AS WE KONW 3 IS THE BEST POINT

```python
plt.plot(k_range,wcss,marker='o')
```

```python
kmeans_final=KMeans(n_clusters=3,random_state=42)
```

```python
cluster_labels=kmeans_final.fit_predict(x_scaled)
```

```python
df['cluster_labels']=cluster_labels
```

```python
sns.scatterplot(x=df['features1'],y=df['features2'],hue=df['cluster_labels'],palette='rainbow')
```
