```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
df["sex"] = le.fit_transform(df["sex"])
df["smoker"]=le.fit_transform(df["smoker"])
df["region"]=le.fit_transform(df["region"])


```

```python
x=df[['age']]
y=df[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
y_predict1=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
# prediction on traning data
y_predict_train=model.predict(x_train)
mse_train1=mean_squared_error(y_train, y_predict_train)

rmse_train1=np.sqrt(mse_train1)
r2_train1=r2_score(y_train,y_predict_train)
print(f"the mse for traning1:{mse_train1}\nthe rmse for traning1 : {rmse_train1}\nthe r2score for traning1: {r2_train1}  ")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
# prediction on traning data

mse_test1=mean_squared_error(y_test, y_predict1)
rmse_test1=np.sqrt(mse_test1)
r2_test1=r2_score(y_test,y_predict1)
print(f"the mse for testing1:{mse_test1}\nthe rmse for testing1 : {rmse_test1}\nthe r2score for testing1: {r2_test1}  ")



```
